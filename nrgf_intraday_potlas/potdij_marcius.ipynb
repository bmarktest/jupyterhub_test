```python
def potdij_szamito(dataframe, reduction=2.85):
    '''Penalty calculating function. Inputs: see description above'''
    df = dataframe.copy()
    df['da_hiba'] = (df['f1']-df['kw']) / 4
    df['elteres_da'] = (np.sign(df['da_hiba']) == np.sign(df['okozoi_elteres_da_total']))
    df['da_share'] = df['elteres_da'] * (df['da_hiba'] / df['okozoi_elteres_da_total'])
    
    df['id_hiba'] = (df['f2']-df['kw']) / 4
    df['elteres_id'] = (np.sign(df['id_hiba']) == np.sign(df['okozoi_elteres_id_total']))
    df['id_share'] = df['elteres_id'] * (df['id_hiba'] / df['okozoi_elteres_id_total'])

    df['nyers_potdij'] = df['kat_koltseg'] * (df['da_share'] + df['id_share']) / 2
    df['csokkentes'] = reduction * df['kw'] / 4
    df['csokkentett_potdij'] = df['nyers_potdij'] - df['csokkentes']
    
    return df
```
